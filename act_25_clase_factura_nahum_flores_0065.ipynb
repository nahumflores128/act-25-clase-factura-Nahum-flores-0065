Se requiere crear un programa en Python utilizando Programación Orientada a Objetos (POO) para digitalizar el control de ventas.

El problema consiste en modelar una clase llamada Factura que permita registrar y organizar la información de dos transacciones comerciales (folio, cliente y monto total). El sistema debe ser capaz de capturar estos datos de forma individual y generar un reporte impreso para cada factura, facilitando así la administración y lectura de los registros financieros.

<img src="https://raw.githubusercontent.com/nahumflores128/act-25-clase-factura-Nahum-flores-0065/refs/heads/main/infoclasefactura.png" width="600">

```python
class Factura:
    def __init__(self, folio, cliente, total):
        self.folio = folio
        self.cliente = cliente
        self.total = total

    # Función 1: Mostrar los datos de la factura
    def mostrardatos(self):
        print(f"\n>>> DETALLES DE FACTURA #{self.folio} <<<")
        print(f"Cliente: {self.cliente}")
        print(f"Total a pagar: ${self.total}")
        print("-" * 30)

    # Función 2: Una función extra para simular lectura o confirmación
    # (En Python los datos suelen pedirse fuera o mediante un método)
    def leerdatos(self):
        print(f"Procesando la factura del cliente {self.cliente}...")
        print("Datos cargados correctamente en el sistema.")

# --- Entrada de datos para la Factura 1 ---
print("--- REGISTRO FACTURA 1 ---")
f1_folio = input("Número de folio: ")
f1_cliente = input("Nombre del cliente: ")
f1_total = float(input("Monto total: "))

# --- Entrada de datos para la Factura 2 ---
print("\n--- REGISTRO FACTURA 2 ---")
f2_folio = input("Número de folio: ")
f2_cliente = input("Nombre del cliente: ")
f2_total = float(input("Monto total: "))

# Creamos los dos objetos (instancias de la clase)
factura1 = Factura(f1_folio, f1_cliente, f1_total)
factura2 = Factura(f2_folio, f2_cliente, f2_total)

# --- Ejecución de Funciones ---

print("\n" + "="*50)
print(f" SISTEMA DE FACTURACIÓN ")
print("="*50)

# Acciones para factura 1
factura1.leerdatos()
factura1.mostrardatos()

# Acciones para factura 2
factura2.leerdatos()
factura2.mostrardatos()

print("\n" + "="*50)
print("PROCESO TERMINADO")

print("Progama realizado por Nahum flores NL = 18")
```
